# 1. Scrape the details of most viewed videos on YouTube from Wikipedia:

Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/

You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
driver=webdriver.Chrome("chromedriver.exe")
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(5)

#to find top videos
top_vid=driver.find_element_by_xpath('//li[@class="toclevel-1 tocsection-2"]//a')
top_vid.click()
time.sleep(1)

#creating list
rank = []
name = []
artist = []
upload_date = []
views = []                                                     

for i in range (1, 31): #using for loop to scrape all data at onc time
    table='//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody//tr['
    rank_path=driver.find_element_by_xpath(table + str(i) + ']//td[1]')
    rank.append(rank_path.text)
    
    name_path=driver.find_element_by_xpath(table + str(i) + ']//td[2]')
    name.append(name_path.text)
    
    artist_path=driver.find_element_by_xpath(table + str(i) + ']//td[3]')
    artist.append(artist_path.text)
    
    views_path=driver.find_element_by_xpath(table + str(i) + ']//td[4]')
    views.append(views_path.text)
    
    upload_date_path=driver.find_element_by_xpath(table + str(i) + ']//td[5]')
    upload_date.append(upload_date_path.text)
     
#creating dataframe   
top_videos=pd.DataFrame({})
top_videos['Rank']=rank
top_videos['Name']=name
top_videos['Artist']=artist
top_videos['Views']=views
top_videos['Upload Date']=upload_date
top_videos.set_index('Rank', inplace = True)
top_videos

#removing inverted comma "--" and [int] numbers from end of name
top_videos['Name'] = top_videos['Name'].apply(lambda x: x[1:-5])        
top_videos
```

```python
#creating list
match_title=[]
series = []
place = []
date = []
time = []  

title_tag=driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]')
for i in title_tag:
    if i.text is None :
        match_title.append("--") 
    else:
        match_title.append(i.text)
        
series_tag=driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]')
for i in series_tag:
    if i.text is None :
        series.append("--") 
    else:
        series.append(i.text)
    
place_tag=driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]//span')
for i in place_tag:
    if i.text is None :
        place.append("--") 
    else:
        place.append(i.text)
    
date_tag=driver.find_elements_by_xpath('//span[@class="fixture__datetime tablet-only"]//strong[1]')
for i in date_tag:
    if i.text is None :
        date.append("--") 
    else:
        date.append(i.text)
    
time_tag=driver.find_elements_by_xpath('//span[@class="fixture__datetime tablet-only"]')
for i in time_tag:
    if i.text is None :
        time.append("--") 
    else:
        time.append(i.text)
        
        
print(len(match_title),len(series),len(place),len(date),len(time))
    
#creating dataframe
match_detail=pd.DataFrame({})                                              
match_detail['Title']=match_title
match_detail['Series']=series
match_detail['Place']=place
match_detail['Date']=date
match_detail['Time']=time
match_detail
```

# 8. Scrape the details of Highest selling novels.

You have to find the following details:
    
A) Book name

B) Author name

C) Volumes sold

B) Year span

D) Run time

# 10. Details of Datasets from UCI machine learning repositories.

Url = https://archive.ics.uci.edu/
    
You have to find the following details:
    
A) Dataset name

B) Data type

C) Task

D) Attribute type

E) No of instances

F) No of attribute

G) Year

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
#to explore chrome
driver=webdriver.Chrome("chromedriver.exe")
url10 = 'https://archive.ics.uci.edu/'
driver.get(url10)
time.sleep(3) 

#to click on "view all dataset"
ds=driver.find_element_by_xpath('//font[@size="3"]')
ds.click()
time.sleep(3)

#create list
name=[]
data_type=[]
task=[]
att_type=[]
inst=[]
no_att=[]
year=[]

#to get all table data together
all_data_tag=driver.find_element_by_xpath('//table[@width="100%"]//p[@class="big"]//b')
all_data=int(all_data_tag.text)+2

#creating a list in which each element will contain all the column
for i in range (2,all_data):          
    
    table_tag=driver.find_elements_by_xpath('//table[@border="1"]//tr['+str(i)+']//p')
    name.append(table_tag[0].text)
    data_type.append(table_tag[1].text)
    task.append(table_tag[2].text)
    att_type.append(table_tag[3].text)
    inst.append(table_tag[4].text)
    no_att.append(table_tag[5].text)
    year.append(table_tag[6].text)
    
#creating dataframe
DS_UCI=pd.DataFrame({})                                             
DS_UCI['Dataset name']=name
DS_UCI['Data type']=data_type
DS_UCI['Task']=task
DS_UCI['Attribute type']=att_type
DS_UCI['No of instances']=inst
DS_UCI['No of attribute']=no_att
DS_UCI['Year']=year
DS_UCI

```
