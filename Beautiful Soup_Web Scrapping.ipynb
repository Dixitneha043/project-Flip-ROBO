# 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
page= requests.get("https://bookpage.com/reviews/")
page
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
soup
```

```python
import pandas as pd
```

```python
#Extracting Next psge urls tags
url_tags=soup.find_all('div', class_="pagination")
url_tags
```

```python
urls=[] #empty list
for i in url_tags:
    for j in i.find_all('a'):
            urls.append(j['href'])
urls
```

```python
#Scrpe book name from multiple pages
book_name=[] #empty list
for url in urls:
    book=requests.get('https://bookpage.com'+url)
    soup=BeautifulSoup(book.content,'html.parser')
    name=soup.find_all('h4',class_='italic') #scrape boosk name
    for i in name:
        book_name.append(i.text)
book_name[:7] #here we are scraping ':7' because it shows top two line also
```

```python
len(book_name)
```

```python
#Scrpe book author name from multiple pages
author_name=[] #empty list
for url in urls:
    book=requests.get('https://bookpage.com'+url)
    soup=BeautifulSoup(book.content,'html.parser')
    author=soup.find_all('p',class_='sans bold') #scrape books author name
    for i in author:
        author_name.append(i.text)
author_name 
```

```python
len(author_name)
```

```python
#Scrpe book genre from multiple pages

genre_book=[] #empty list
for url in urls:
    book=requests.get('https://bookpage.com'+url)
    soup=BeautifulSoup(book.content,'html.parser')
    genre=soup.find_all('p',class_='genre-links hidden-phone') #scrape books author name
    for i in genre:
        genre_book.append(i.text)
genre_book 
```

```python
len(genre_book)
```

```python
#Scrpe book reviews from multiple pages

book_reviews=[] #empty list
for url in urls:
    book=requests.get('https://bookpage.com'+url)
    soup=BeautifulSoup(book.content,'html.parser')
    reviews=soup.find_all('div',class_='read-full') #scrape books author name
    for i in reviews:
        book_reviews.append(i.text)
genre_book 
```

```python
len(book_reviews)
```

```python
#creating python program to create a list of book name, reviews, author name and genre

Book_detail=pd.DataFrame({})
Book_detail['--------Review-------']=book_reviews
Book_detail['-------Book Name--------']=book_name
Book_detail['----------Author Name-----------']=author_name
Book_detail['----------Genre----------']=genre_book
```

```python
#Scrpe mobiles name from multiple pages
mobiles_name=[] #empty list
for url in urls:
    name=soup.find_all('a',class_='a-link-normal a-text-normal') #scrape mobile name
    for i in name:
        mobiles_name.append(i.text)
mobiles_name
```

```python
len(mobiles_name)
```

```python
len(mobiles_price)
```

```python
#Scrpe mobiles rating from multiple pages
mobiles_rating=[] #empty list
for url in urls:
    rating=soup.find_all('a',class_='a-popover-trigger a-declarative') #scrape mobile rating
    for i in rating:
        mobiles_rating.append(i.text)
mobiles_rating
```

```python
len(mobiles_rating)
```
