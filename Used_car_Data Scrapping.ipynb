```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

from selenium.webdriver.common.keys import Keys
#Importing requests
import requests

# importing regex
import re
```

```python
car_urls=[]
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)
car_urls=[]

url = "https://www.cardekho.com/used-cars+in+delhi-ncr"
driver.get(url)
time.sleep(3)
driver.maximize_window()
```

```python
time.sleep(3)
previous_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
   
    driver.execute_script("window.scrollBy(0,document.documentElement.scrollHeight);")
    time.sleep(3)
    
    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    time.sleep(3)
    if new_height == previous_height:
        break
    previous_height = new_height
```

```python

urls = driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7 carsName']//a")
for i in urls:
    car_urls.append(i.get_attribute('href'))
car_urls
```

```python
len(car_urls)
```

```python
car_name_=[]
make_year=[]
Reg_year=[]
fule=[]
km_driven=[]
Engine_Displacement=[]
No_Of_Owner=[]
RTO=[]
Transmission=[]
Insurance_Type=[]
Range_price=[]
Avg_market_price=[]
New_Car_Road_Price=[]
Mileage=[]
Seats=[]
Color=[]
Engine=[]
Max_Power=[]
Torque=[]

from selenium.common.exceptions import NoSuchElementException

for p in car_urls:
    driver.get(p)
    time.sleep(3)
    
    try:
        C_name = driver.find_element_by_xpath('//div[@class="paddingBorder clearfix"]//h1')      
        car_name_.append(C_name.text)
    except NoSuchElementException:
        car_name_.append('-')
    
    try:
        C_year = driver.find_element_by_xpath("//*[@id='rf01']/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[1]/div/div[2]")
        make_year.append(C_year.text)
    except NoSuchElementException:
        make_year.append('-')
    
    try:
        C_Reg_year= driver.find_element_by_xpath("//*[@id='rf01']/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[2]/div/div[2]")
        Reg_year.append(C_Reg_year.text)
    except NoSuchElementException:
        Reg_year.append("-")
    
    try:
        C_fule= driver.find_element_by_xpath("//*[@id='rf01']/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[3]/div/div[2]")
        fule.append(C_fule.text)
    except NoSuchElementException:
        fule.append("-")
    
    try:
        C_driven= driver.find_element_by_xpath("//*[@id='rf01']/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[4]/div/div[2]")
        km_driven.append(C_driven.text)
    except NoSuchElementException:
        km_driven.append("-")
        
    try:
        C_Engine_Displacement= driver.find_element_by_xpath("//*[@id='rf01']/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[5]/div/div[2]")
        Engine_Displacement.append(C_Engine_Displacement.text)
    except NoSuchElementException:
        Engine_Displacement.append("-")
        
    try:
        C_No_Of_Owner= driver.find_element_by_xpath("//*[@id='rf01']/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[6]/div/div[2]")
        No_Of_Owner.append(C_No_Of_Owner.text)
    except NoSuchElementException:
        No_Of_Owner.append("-")
        
    try:
        C_RTO= driver.find_element_by_xpath("//*[@id='rf01']/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[7]/div/div[2]")
        RTO.append(C_RTO.text)
    except NoSuchElementException:
        RTO.append("-")
    
    try:
        C_Transmission= driver.find_element_by_xpath("//*[@id='rf01']/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[8]/div/div[2]")
        Transmission.append(C_Transmission.text)
    except NoSuchElementException:
        Transmission.append("-")

    try:
        C_Insurance_Type= driver.find_element_by_xpath("//*[@id='rf01']/div[1]/div/div/main/div/div/div[1]/div[3]/div/div/div/ul/li[9]/div/div[2]")
        Insurance_Type.append(C_Insurance_Type.text)
    except NoSuchElementException:
        Insurance_Type.append("-")
    
    try:
        C_range_price= driver.find_element_by_xpath("//div[@class='pricerangeDiff gsc_col-xs-4 gsc_col-lg-3 posR movable paddingRight']//div")
        Range_price.append(C_range_price.text)
    except NoSuchElementException:
        Range_price.append("-")
        
    try:
        C_avg_market_price= driver.find_element_by_xpath("//div[@class='pricerangeDiff gsc_col-xs-4 gsc_col-lg-5 posR text-center paddingLR']//div")
        Avg_market_price.append(C_avg_market_price.text)
    except NoSuchElementException:
        Avg_market_price.append("-")
        
    try:
        C_Road_Price= driver.find_element_by_xpath("//div[@class='pricerangeDiff gsc_col-xs-4 text-right paddingleft posR green']//div")
        New_Car_Road_Price.append(C_Road_Price.text)
    except NoSuchElementException:
        New_Car_Road_Price.append("-")
        
    try:
        C_Mileage= driver.find_element_by_xpath("//*[@id='specification-wdght']/div/div/div/ul/li[1]/div/div[2]")
        Mileage.append(C_Mileage.text)
    except NoSuchElementException:
        Mileage.append("-")

    #time.sleep(3)
    #driver.execute_script("window.scrollBy(0,1000)","")
    #button=driver.find_element_by_id("specification-wdght")
    #button.click()
    #time.sleep(3)
        
    try:
        C_Seats= driver.find_element_by_xpath("//*[@id='specification-wdght']/div/div/div/ul/li[5]/div/div[2]")
        Seats.append(C_Seats.text)
    except NoSuchElementException:
        Seats.append("-")        
        
    try:
        C_Color= driver.find_element_by_xpath("//*[@id='specification-wdght']/div/div/div/ul/li[6]/div/div[2]")
        Color.append(C_Color.text)
    except NoSuchElementException:
        Color.append("-")  

   
    try:
        C_Engine= driver.find_element_by_xpath("//*[@id='specification-wdght']/div/div/div/ul/li[2]/div/div[2]")
        Engine.append(C_Engine.text)
    except NoSuchElementException:
        Engine.append("-")
        

    try:
        C_Max_Power= driver.find_element_by_xpath("//*[@id='specification-wdght']/div/div/div/ul/li[3]/div/div[2]")
        Max_Power.append(C_Max_Power.text)
    except NoSuchElementException:
        Max_Power.append("-")  

    try:
        C_Torque= driver.find_element_by_xpath("//*[@id='specification-wdght']/div/div/div/ul/li[4]/div/div[2]")
        Torque.append(C_Torque.text)
    except NoSuchElementException:
        Torque.append("-")
```

```python
car_details=pd.DataFrame({"Car Name": car_name_,"Making Year":make_year,"Registration year":Reg_year,"Fule":fule,"km_driven":km_driven,"Engine_Displacement":Engine_Displacement, "No_Of_Owner":No_Of_Owner,"RTO":RTO,"Transmission":Transmission,"Insurance_Type":Insurance_Type,"Range_price":Range_price,"Avg_market_price":Avg_market_price,"New_Car_Road_Price":New_Car_Road_Price,"Mileage":Mileage,"Seats":Seats,"Color":Color,"Engine":Engine,"Max_Power":Max_Power,"Torque":Torque})
```
